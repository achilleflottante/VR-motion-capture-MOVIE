```python
df = pd.read_csv('16.csv')
mes_avant = "Salut,\nTu as "
mes_apres = "€ de dettes à la biero. Pourrais-tu, stp, rembourser via ce lien (https://collecte.io/dettes-biero-740792/fr), ou mardi prochain en biero, en espèce ou grâce à Lydia. N'hésites pas à mettre plus que ce qu'il faut pour rembourser tes dettes, car les crédits sont censés être exceptionnels.\nMerci par avance\nValentin "
```

```python
for id in df.index:
    if df.loc[id, "num_bool"]==True :
        message = mes_avant + str(df.loc[id, "solde"])[1:] + mes_apres
        num = "+33" + df.loc[id, "num"][1:]
        pwk.sendwhatmsg_instantly(num, message)
```
