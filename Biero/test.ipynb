```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python

def construction(annee):
    session = requests.Session()
    url = 'http://10.20.1.30/octogone/login.php'
    url2 = 'http://10.20.1.30/octogone/recherche.php?year=' + annee + '&first_letter=a&second_letter=z'
    log = {"login" : "23allard", "mdp" : mdp }
    response = session.post(url, data=log)
    doc = session.get(url2)
    soup = BeautifulSoup(doc.text, 'html.parser')
    clients = {"id": [], "solde" : [], "infos_sup" : []}
    client_elements = soup.find_all(class_="client")
    for client_element in client_elements:
        pid = client_element.find('h3').text
        if pid != 'Suivants':
            infos = client_element.find('p').text
            if infos[-12:] == 'Biero\n\t\t\t\t\t\t':
                infos = infos[:-22]
            solde = infos [-7:-1]
            if solde[0] == ' ':
                solde = solde[1:]
            if solde[0] == ':':
                solde = solde[1:]
            clients["id"].append(pid)
            clients["solde"].append(float(solde))
            clients["infos_sup"].append(infos[4:])
    df = pd.DataFrame.from_dict(clients)
    df = df.set_index("id")
    session.close()
    return df
```
