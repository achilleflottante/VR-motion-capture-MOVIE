```python
import pywhatkit as pwk
import pandas as pd

#ATTENTION IL Y A UN MAUVAIS NUMERO
```

```python
premPart = "Salut, C'est Valentin, le trésorier de la Biero. Tu as plus de 6€ de dettes ("
secPart = " € de dettes précisemment) donc ça serait bien si tu pouvais rembourser. Tu peux soit rembourser via ce lien :https://collecte.io/dettes-biero-740792/fr, soit payer en liquide.\n Merci par avance \n Valentin"
```

```python
df = pd.read_csv('18_10_2024.csv')
df = df[df['year'] > 21]
df
```

```python
df = df[df['Solde']<=-6]
df
```

```python
len(df)
```

```python
for id in df.index:
    if df.loc[id, "num_bool"]==True :
        message = premPart + str(df.loc[id, "Solde"])[1:] + secPart
        num = "+33" + df.loc[id, "num"][1:]
        pwk.sendwhatmsg_instantly(num, message)
```
